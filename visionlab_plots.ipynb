```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
# x=[50]*len(ang)
avg=np.asarray([10.437, 10.188, 10.117, 10.131, 10.049, 9.958, 10.058])
std=np.asarray([0.0105, 0.0173, 0.0203, 0.0194, 0.0192, 0.0190, 0.0171])**2
# x=[50]*len(avg)
x=np.arange(50,50*len(avg)+1,50)
# plt.plot(x,avg)
# plt.xlabel('samples')
# plt.ylabel('frequency')

f, (ax1, ax2) = plt.subplots(1, 2, sharey=False)
f.tight_layout()
ax1.plot(x, avg)
ax1.set_title('frequency mean with nodelets')
ax1.set_xlabel('samples')
ax1.set_ylabel('frequency')
ax1.set_ylim(8,11)
ax2.plot(x, std)
ax2.set_title('frequency variance vs nodelets')
ax2.set_xlabel('samples')
ax2.set_ylabel('frequency')
# ax2.set_ylim(0,0.03)
ax2.set_ylim(0,0.01)
plt.show()
```
